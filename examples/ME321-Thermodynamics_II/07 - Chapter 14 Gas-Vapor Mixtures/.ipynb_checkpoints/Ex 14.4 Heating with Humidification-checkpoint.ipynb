## Problem Statement
Using a psychrometric chart determine the specific humidity, $\omega$, relative humidity, $\phi$, and specific enthalpy, $h$, for atmospheric air with $T_{db}=20^\circ\mathrm{C}$ and $T_{wb}=15^\circ\mathrm{C}$


## Solution


### Python Initialization
We'll start by importing the libraries we will use for our analysis and initializing dictionaries to hold the properties we will be usings.

```python
from thermoJFM.templates.humidair_default import *

psych = humidair.psychrometric_chart()
```

### Given

```python
p = Quantity(1,'atm')
T[1] = Quantity(20,'degC')
T_wb[1] = Quantity(15,'degC')
states.fix(1,humidair)

psych.plot_state(states[1])
psych.show()
```

#### Properties

```python
# Humidity Ratio: y-axis
Tmin = humidair.T(rel_hum=1,omega=omega[1])
#psych._plot_straight_line(x1=Tmin,x2=psych.Tmax,y1=omega[1],y2=omega[1])
psych._plot_iso_wrapper('w',omega[1],linewidth=2,linestyle='-',pos=0.1,label=f"$\omega_1={omega[1].magnitude:.5f}\ \mathrm{{{psych.props.format_units(f'{omega[1].units}')}}}$")

# Relative_humidity
psych._plot_iso_wrapper('rel_hum',rel_hum[1],linewidth=2,linestyle='-',pos=0.1,label=f'$\phi_1={rel_hum[1].magnitude*100:.2f}$%')

# Enthalpy
psych._plot_iso_wrapper('h',h[1],linewidth=2,linestyle='-',pos=0.1,label=f"$h_1={h[1].magnitude:.2f}\ \mathrm{{psych.props.format_units(f'{h[1].units}')}}$")

psych.show()

display.Calculations(locals(),comments=True);
```

```python
omega[1]
```

### Wet Bulb Specific Humidity (Saturated)

```python
omega_wb = (0.622*p_g_wb)/(p-p_g_wb)
omega_wb = omega_wb.to('')

display.Calculations(locals(),comments=True);
```

### Dry Bulb Specific Humidity

```python
omega = ( c_p*(T_wb-T_db) + omega_wb*h_fg_wb )/( h_g_db-h_f_wb )

display.Calculations(locals(),comments=True);
```

### Dry Bulb Specific Humidity

```python
phi = omega*p/((0.622+omega)*p_g_db)
phi = phi.to('') # hide

display.Calculations(locals(),comments=True);
```

### Dry Bulb Enthalpy

```python
T_db_C = Quantity(T_db.magnitude,'delta_degC')
h = c_p*T_db_C + omega*h_g_db

display.Calculations(locals(),comments=True);
```
